```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import statsmodels
```

```python
pd.read_csv("/home/ydong/dataSets/mini/absP/0.csv",index_col = "GeneID")
```

```python
dfList = []
for i in range(0,377):
    print(i)
    currDF = pd.read_csv("/home/ydong/dataSets/mini/absP/"+str(i)+".csv",index_col = "GeneID")
    dfList.append(currDF)
```

```python
absP = pd.concat(dfList,axis = 1)
```

```python
absP = absP.transpose()
```

```python
absP
```

```python
correctedP = absP.copy() 
for i in absP.columns:
    currP = absP[i]
    rej,corP = statsmodels.stats.multitest.fdrcorrection(currP)
    correctedP[i] = corP
```

```python
correctedP.to_csv("correctedPVals.csv")
```
